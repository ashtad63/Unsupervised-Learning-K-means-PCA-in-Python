## Iris Dataset â€“ K-Means and PCA 

```python
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

### Step A

```python
iris_data = pd.read_csv('iris.csv')
iris_data.head() #10
```

```python
iris_data.tail() #10
```

```python
iris_data.shape
```

```python
iris_data.dtypes
```

```python
iris_data.groupby(['Species']).describe().T
```

### Step B

```python
iris_corr = iris_data.corr()
iris_corr
```

```python
grps = ['NegStrong','NegMedium','Weak', 'PosMedium', 'PosStrong']
cols = ['SepalLength','SepalWidth', 'PetalLength','PetalWidth']
for col in cols:
    iris_corr[col] = pd.cut(iris_corr[col], (-1,-0.8,-0.6,0.6,0.8,1), labels = grps)
iris_corr
```

```python
sns.pairplot(iris_data, hue='Species', height=2.5);
```

```python
spec_cat = pd.factorize (iris_data['Species'])
Y_iris = spec_cat[0]
iris_names = ['Sepal Length','Sepal Width', 'Petal Length','Petal Width']
Y_iris
```

```python
x = 0  # Column Sepal Length
y = 1  # Column Sepal Width
plt.scatter(iris_data.iloc[:,x], 
            iris_data.iloc[:,y], 
            c = Y_iris, s = 50, 
            cmap = 'viridis');
plt.xlabel (iris_names[x]);
plt.ylabel (iris_names[y]);
```

```python
x = 2  # Column Petal Length
y = 3  # Column Petal Width
plt.scatter (iris_data.iloc[:,x],
            iris_data.iloc[:, y], 
            c = Y_iris, s = 50, 
            cmap = 'viridis');
plt.xlabel (iris_names[x]);
plt.ylabel (iris_names[y]);
```

### Step C

```python
X_iris = iris_data.drop ('Species', axis=1)
# or X_iris = iris_data.iloc[:, :4]
```

```python
kmeans = KMeans (n_clusters=3, random_state = 1)
kmeans.fit (X_iris)
y_model = kmeans.predict (X_iris)
y_model
```

```python
y_test = np.choose (Y_iris,[1,0,2]).astype (int)
y_test
```

```python
acc_score = accuracy_score (y_test, y_model)
acc_score
```

```python
plt.scatter(X_iris.iloc[:,0], X_iris.iloc[:,1], c = y_model, s = 50, cmap = 'viridis');
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c = 'black', s = 100, alpha = 0.5);
```

```python
mat = confusion_matrix (y_test, y_model)
sns.heatmap (mat, square=True, annot=True, cbar=False);
plt.xlabel('Predicted Values');
plt.ylabel('Actual Values');
```

### Step D

```python
model = PCA (n_components=2) 
model.fit (X_iris) 
X_2D = model.transform(X_iris)
```

```python
for label, color, i in zip(('Setosa','Versicolor','Virginica'), ('blue','red','green'), (0, 1, 2)):
    plt.scatter (X_2D[Y_iris == i, 0], 
                 X_2D[Y_iris == i, 1], 
                 label=label, color=color);
plt.xlabel('Principal Component 1');
plt.ylabel('Principal Component 2');
plt.legend(loc='lower right');
```

### Step E

```python
X_std_iris = StandardScaler().fit_transform(X_iris)
X_std_iris
```

```python
kmeans = KMeans(n_clusters=3, random_state = 1)
kmeans.fit(X_std_iris)
y_model = kmeans.predict(X_std_iris)
```

```python
acc_score = accuracy_score(y_test, y_model)
acc_score
```

```python
plt.scatter (X_std_iris[:,0], X_std_iris[:,1], c = y_model, s = 50, cmap = 'viridis')
centers = kmeans.cluster_centers_
plt.scatter (centers[:,0], centers[:,1], c = 'black', s = 100, alpha = 0.5);
```

```python
mat = confusion_matrix (y_test, y_model)
sns.heatmap (mat, square=True, annot=True, cbar=False)
plt.xlabel ('Predicted Values')
plt.ylabel ('Actual Values');
```

### Step F

```python
X_norm_iris = MinMaxScaler().fit_transform(X_iris)
X_norm_iris
```

```python
kmeans = KMeans(n_clusters=3, random_state = 1)
kmeans.fit(X_norm_iris)
y_model = kmeans.predict(X_norm_iris)
```

```python
acc_score = accuracy_score (y_test, y_model)
acc_score
```

```python
plt.scatter (X_norm_iris[:,0], X_norm_iris[:,1], c = y_model, s = 50, cmap = 'viridis')
centers = kmeans.cluster_centers_
plt.scatter (centers[:,0], centers[:,1], c = 'black', s = 100, alpha = 0.5);
```

```python
mat = confusion_matrix (y_test, y_model)
sns.heatmap (mat, square=True, annot=True, cbar=False)
plt.xlabel ('Predicted Values')
plt.ylabel ('Actual Values');
```
